```python
#Bar Graph
import matplotlib.pyplot as plt
x=df.quality
y=df.pH
plt.bar(x,y,label="wine quality",color="red")
plt.xlabel("quality")
plt.ylabel("pH")
plt.title("wine quality vs pH")
plt.legend()
```

```python
#creating horizontal bar graphs
import matplotlib.pyplot as plt
x=[1001,1003,1005,1006,1009]
y=[10000,23000.50,18000,16000,9999.9]
x1=[1002,1004,1010,1008,1014]
y1=[5000,6000,4000,9000,10000]
plt.barh(x,y,label='Sales Department',color="red")
plt.barh(x1,y1,label="production department",color="blue")
plt.xlabel("Empids")
plt.ylabel('Salaries')
plt.legend()
plt.title("Sales vs Production")
```

```python
import matplotlib.pyplot as plt
sugar_men=[113,85,90,150,149,88,93,115,135,80]
sugar_women=[103,80,85,110,112,88,93,125,132,70]
bins=[80,100,125,150]
(n,bins,patches)=plt.hist([sugar_men,sugar_women],bins,rwidth=0.90,color=["green","orange"],label=['men','women'])
plt.xlabel('sugar levels')
plt.ylabel('number of people')
plt.title('sugar levels')
plt.legend()
```

```python
import matplotlib.pyplot as plt
sugar_men=[113,85,90,150,149,88,93,115,135,80]
sugar_women=[103,80,85,110,112,88,93,125,132,70]
bins=[80,100,125,150]
(n,bins,patches)=plt.hist([sugar_men,sugar_women],bins,rwidth=0.90,color=["green","orange"],label=['men','women'],orientation='horizontal')
plt.xlabel('sugar levels')
plt.ylabel('number of people')
plt.title('sugar levels')
plt.legend()
```

```python
import matplotlib.pyplot as plt
labels = ['Apples', 'Bananas', 'Oranges', 'Grapes']
sizes = [30, 25, 20, 25]
colors = ['red', 'yellow', 'orange', 'purple']
explode = (0.1, 0, 0, 0)
plt.pie(sizes, explode=explode, labels=labels, colors=colors,autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.title('Fruit Distribution')
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 5, 3]
plt.plot(x, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Graph")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 1, 5, 3]
y2 = [1, 3, 2, 4, 2]
plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Double Line Graph')
plt.legend()
plt.show()
```
